```python
import pandas as pd # used for import pandas lib
```

```python
def matchingFunction(data, column):# funct data=df , column = Skills
  data=preprocess(data, column)# funct data=df , column = Skills
  matchml(data,column) #funct data=df , column = Skills
  matchweb(data,column)#funct data=df , column = Skills
  applyDesignation(data, column)#funct data=df , column = Skills

def preprocess(data, column):
  data[column] = data[column].astype(str).apply(lambda x: x.replace(" ","")) #used for replace " " with ""
  data[column] = data[column].astype(str).apply(lambda x: x.replace("'",""))#used for replace "'" with ""
  data[column] = data[column].apply(lambda x: x[1:-1].split(','))#used for split list and add , between two element
  return data # return pre process data

def matchml(data, column):#
  data[column+'_Matched_ml']=data[column].apply(Functionml) # create data frame column name = Skills_matched_ml and 
   #then data[column]=df['Skills'] goes as list parameter to functionml  

def matchweb(data, column):
  data[column+'_Matched_web']=data[column].apply(Functionweb)# create data frame column name = Skills_matched_web and
  #then data[column]=df['Skills'] goes as list parameter to functionweb
    
    
def Functionweb(x):
    
  df2=pd.DataFrame({'web':['django','html','css','nodejs','reactjs','flask','laravel'],
  'ml':['nosql','pandas','numpy','matplotlib','seaborn','scipy','na']})# required df2 for both web and ml this skills 
   # classify 
  l=df2['web'].astype(str).values  # convert dataframe of web column into list
  a = set(x) # convert the list into set 
  b = set(l) # convert list into set

  if (len(a.intersection(b))>0): # intersection of set check value 
    return list(a.intersection(b))# convert into list and send the intersect elements
  else:
    return []# otherwise return [] nothing

def Functionml(x):
  df2=pd.DataFrame({'web':['django','html','css','nodejs','reactjs','flask','laravel'],
  'ml':['nosql','pandas','numpy','matplotlib','seaborn','scipy','na']})# required df2 for both web and ml this skills 
   # classify 

  l=df2['ml'].astype(str).values  # convert dataframe of web column into list
  a = set(x) # convert the list into set 
  b = set(l) # convert list into set
  if (len(a.intersection(b))>0):# intersection of set check value 
    return list(a.intersection(b))# convert into list and send the intersect elements
  else:
    return []# otherwise return [] nothing


def applyDesignation(data, column):
  data['Designation']='Not Fitting required skills'# define a column in df dataframe name 'Designation' with initial 
      # 'Not Fitting required skills' as first element

  data['Skills_Matched'] = 0  # define a column in df dataframe name 'Skills_Matched' with initial 
      # 0 as first element
  
  for x in range(0,len(data[column])): # loop 0 to 49
    m=len(data[column+'_Matched_web'][x]) # iterate 0 to 49 column of "Skills_Matched_web" and finds the number of elements
    # present in a particular column 
    
    n=len(data[column+'_Matched_ml'][x])# iterate 0 to 49 column of "Skills_Matched_ml" and finds the number of elements
    # present in a particular column 
    
    if m>=3: # if it is > 3 
        
      if m>n:  # if the value is greater than other column value at the same column of both
        data['Designation'][x]='Web_Developer' # put Web_Developer in Designation column
        data['Skills_Matched'][x]=m # number of skill matched in the column of Skills_Matched
    if n>=3:
      if m<n: # if the value is greater than other column value at the same column of both
        data['Designation'][x]='Machine_learning_Developer' # put Machine_learning_Developer in Designation column
        data['Skills_Matched'][x]=n # number of skill matched in the column of Skills_Matched
```

```python
df=pd.read_excel('skill_data.xlsx')# read given excel file and put in df dataframe
matchingFunction(df, 'Skills') # function call and df as data and Skills as column send
```
